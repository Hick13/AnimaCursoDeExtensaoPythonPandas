```python
# Filtrar apenas por uma coluna
display(combustiveis_df['Revenda'])
```

```python
#Aqui criamos um novo dataframe apenas com as colunas que eu quero...
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']]
display(ca_df)
```

```python
# Exibe a 4º linha. "sempre começa pelo numero 0"
display(ca_df.loc[3])
```

```python
# Exibir da 10º linha até a 20º linha
display(ca_df.loc[9:19])
```

```python
# Criar uma dataframe gas_df contendo
# apenas as 4 colunas (Revenda, Municipio, Produto, Valor de Venda)
# Somente com combustível sendo GASOLINA e exibir a tela
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

display(gas_df['Valor de Venda'].max())
```

```python
#Exibir o posto e municipio da gasolina mais cara do  país
display(gas_df[['Revenda','Municipio', 'Valor de Venda']].max())
```

```python
#Dataframe.loc[] com múltiplas condições para filtragem
#Quais são os preços, postos que vendem ETANOL na minha cidade (INDAIATUBA)?
#ordenando do menor para o maior valor de venda
etanol_indaiatuba_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'INDAIATUBA')]
display(etanol_indaiatuba_df.sort_values(by='Valor de Venda'))
```

```python
#Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADAdo Bairro BELA VISTA em SÃO PAULO
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & (combustiveis_df['Municipio'] == 'SAO PAULO') & (combustiveis_df['Produto'] == 'GASOLINA')])
```

```python
# Mesmo metodo acima mas, utilizando o ou "|"
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') &
                            (combustiveis_df['Municipio'] == 'SAO PAULO') &
                            ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')), 
                            ['Valor de Venda']].mean())
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do Bairro MOOCA em SÃO PAULO?
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            (combustiveis_df['Produto'].isin(["GASOLINA", "GASOLINA ADITIVADA"])), 
                            ['Valor de Venda']].mean())
```

```python
# Como mostrar média de valor de venda POR COMBUSTÍVEL?

# 'groupby' faz agrupamento de informações
# 'Roun' delimita as casas númerais

media_por_combustivel = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_por_combustivel)
```
